```python
with open("input.txt", "r", encoding="utf-8") as f:
    input_data = f.read().strip()
```

```python
def transpose(data: str) -> list:
    return ["".join(x) for x in zip_longest(*data.split("\n"), fillvalue=" ")]
```

```python
def calc_op(op: list) -> int:
    operator = op[0][-1]
    op[0] = op[0][:-1]
    numbers = [int(item) for item in op if item.strip()]
    return sum(numbers) if operator == "+" else math.prod(numbers)
```

```python
gp = "\n".join(row.strip() for row in transpose(input_data)).split("\n\n")

res = sum(calc_op(group.splitlines()) for group in gp if group.strip())
```
