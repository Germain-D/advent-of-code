```python
def parse_input(input_file):
    with open(input_file, 'r') as file:
        return file.read().splitlines()
```

```python
inputs = parse_input("input1.txt")
```

```python
pos = 50
counter = 0
for input in inputs:
    dist_to0 = abs(pos)
    nombre = (int(input[1:]))% 100
    if input.startswith("L"):
        pos -= nombre

        if pos < 0 and pos != 0:
            pos += 100

    elif input.startswith("R"):
        pos += nombre
        if pos + nombre > 100 and pos != 0:
            pos -= 100
    if pos == 0:
        counter += 1
    
print(counter)
```
