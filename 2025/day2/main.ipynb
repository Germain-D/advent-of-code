```python
import numpy as np
import pandas as pd
```

```python
def parse_input(input_file):
    with open(input_file, 'r') as file:
        return file.read().strip().split(',')
```

```python
inputs = parse_input("input2.txt")
```

```python
inputs
```

```python
def find_pairs(n):
    n = str(n)
    s = int(len(str(n))/2)
    if n[:s] == n[s:]:
        return True
    return False
```

```python
res = 0
for rge in inputs:
    start, end = map(int, rge.split('-'))
    for i in range(int(start), int(end) + 1):
        if str(i)[0] != '0': 
            if find_pairs(i):
                res += i
print(res)
```
